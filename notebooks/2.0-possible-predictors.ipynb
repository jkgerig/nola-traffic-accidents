```python
# set start and end datetimes for ranges
range_start = pd.Timestamp('2012-01-01 00:00:00')
range_end   = pd.Timestamp('2018-07-01 00:00:00')
```

```python
# create interval indexes
date_range_60_min = pd.interval_range(start=range_start, end=range_end, freq='60min', closed='left')
date_range_1_day = pd.interval_range(start=range_start, end=range_end, freq='D', closed='left')
```

```python
# import weather pickle files
df_weather = pd.read_pickle('../data/interim/weather.pickle')
df_weather.set_index('DATE', inplace=True)
```

```python
# create inclement_weather variable
df_weather['inclement_weather'] = np.where(df_weather.HOURLYPRSENTWEATHERTYPE.isna(), 0, 1)
```

```python
# resample to group by hour and day
df_weather_hour = df_weather[['DAILYPrecip', 'inclement_weather']].resample('60T').max()
df_weather_day =  df_weather[['DAILYPrecip', 'inclement_weather']].resample('D').max()
```

```python
df_hour = pd.DataFrame(index=date_range_60_min)
df_day = pd.DataFrame(index=date_range_1_day)
```
