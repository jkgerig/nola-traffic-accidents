```python
# package imports
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import shapely
```

```python
neighborhoods = gpd.read_file('../../data/raw/Neighborhood_Statistical_Areas/Neighborhood_Statistical_Areas.shp')
neighborhoods = neighborhoods.to_crs(epsg=4326)
neighborhoods = neighborhoods[['OBJECTID', 'GNOCDC_LAB', 'geometry']]
neighborhoods.columns = ['nhood_id', 'nhood', 'geometry']
```

```python
streets = gpd.read_file('../../data/raw/Road_Centerline/geo_export_c02761a8-1d85-477e-a5cf-01b9f22f4d88.shp')

streets_columns = ['joinid',
                   'fromright',
                   'centerline',
                   'roadclass',
                   'fullname',
                   'toleft',
                   'toright',
                   'fullnameab',
                   'objectid',
                   'shape_leng',
                   'fromleft',
                   'geometry']

streets = streets[streets_columns]
```

```python
crs_louisiana = {'proj': 'lcc',
                 'lat_1': 29.3,
                 'lat_2': 30.7,
                 'lat_0': 28.5,
                 'lon_0': -91.33333333333333,
                 'x_0': 1000000,
                 'y_0': 0,
                 'datum': 'NAD83',
                 'units': 'us-ft',
                 'no_defs': True}
```

```python
streets_reference = streets[['joinid', 'roadclass', 'fullname', 'fullnameab', 'objectid', 'shape_leng', 'geometry']]
```

```python
subset = ['joinid', 'roadclass', 'shape_leng', 'geometry']
dissolve_by = ['joinid', 'roadclass']
streets_dissolved = streets[subset].dissolve(by=dissolve_by, aggfunc='sum')
streets_dissolved = streets_dissolved.reset_index(level=[0,1])
```

```python
streets_neighborhoods_join = gpd.sjoin(streets_dissolved, neighborhoods)
```
